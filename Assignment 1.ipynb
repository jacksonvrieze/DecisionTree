```python
import pandas as pd
import statistics as stat
import statsmodels.api as sm
import statsmodels.formula.api as smf
import numpy as np
import seaborn as sns
```

```python

train = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data", names=['sepal_length','sepal_width','petal_length','petal_width','class'])
```

```python

train = pd.DataFrame(train)
print(train)
```

```python
train.describe()
#train.mean(axis=0)
#train['petal_length'].value_counts()
#train['petal_length'].std()
```

```python
boxplot = train.boxplot(column=['petal_length','petal_width','sepal_length','sepal_width'])
boxplot
```

```python
histogram = train.hist(column=['petal_length','petal_width','sepal_length','sepal_width'])
```

```python
# pp_x = sm.ProbPlot(train['a'])
 #pp_y = sm.ProbPlot(train['b'])
sm.qqplot_2samples(train['sepal_length'], train['sepal_width'],xlabel= 'Sepal Width', ylabel= "Sepal Length")
```

```python
sns.set()
# Plot sepal with as a function of sepal_length across days
g = sns.lmplot(x='sepal_length', y='sepal_width', hue='class',
               truncate=True, height=5, data=train)

# Use more informative axis labels than are provided by default
g.set_axis_labels("Sepal length (mm)", "Sepal width (mm)")
```

```python
h = sns.lmplot(x='sepal_length', y='petal_length', hue='class',
               truncate=True, height=5, data=train)

# Use more informative axis labels than are provided by default
h.set_axis_labels("Sepal length (mm)", "Petal length (mm)")
```
